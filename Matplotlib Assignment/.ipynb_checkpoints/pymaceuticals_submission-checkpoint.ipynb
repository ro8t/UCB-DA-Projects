```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug = pd.read_csv(mouse_drug_data_to_load)
clinical = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
main = pd.merge(clinical, mouse_drug, how='left', on=['Mouse ID', 'Mouse ID'])

# Display the data table for preview
main.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
avg_tumor_vol = main[['Drug', 'Timepoint', 'Tumor Volume (mm3)']].groupby(['Drug', 'Timepoint']).mean()
avg_tumor_vol = pd.DataFrame(avg_tumor_vol).reset_index()
avg_tumor_vol.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor_vol = main[['Drug', 'Timepoint', 'Tumor Volume (mm3)']].groupby(['Drug', 'Timepoint']).sem()
sem_tumor_vol = pd.DataFrame(sem_tumor_vol).reset_index()
sem_tumor_vol.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
avg_tumor_vol_pivot = avg_tumor_vol.pivot(index='Timepoint', columns='Drug')['Tumor Volume (mm3)']
sem_tumor_vol_pivot = sem_tumor_vol.pivot(index='Timepoint', columns='Drug')['Tumor Volume (mm3)']

avg_tumor_vol_pivot.head()
#sem_tumor_vol_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(avg_tumor_vol_pivot.index, avg_tumor_vol_pivot['Capomulin'], yerr=sem_tumor_vol_pivot['Capomulin'], color="b", marker="o", linewidth=0.3, ms=4)
#plt.errorbar(avg_tumor_vol_pivot.index, avg_tumor_vol_pivot['Ceftamin'], yerr=sem_tumor_vol_pivot['Ceftamin'], color="g", marker="v", linewidth=0.3, ms=4)
plt.errorbar(avg_tumor_vol_pivot.index, avg_tumor_vol_pivot['Infubinol'], yerr=sem_tumor_vol_pivot['Infubinol'], color="r", marker="^", linewidth=0.3, ms=4)
plt.errorbar(avg_tumor_vol_pivot.index, avg_tumor_vol_pivot['Ketapril'], yerr=sem_tumor_vol_pivot['Ketapril'], color="c", marker="<", linewidth=0.3, ms=4)
#plt.errorbar(avg_tumor_vol_pivot.index, avg_tumor_vol_pivot['Naftisol'], yerr=sem_tumor_vol_pivot['Naftisol'], color="m", marker=">", linewidth=0.3, ms=4)
plt.errorbar(avg_tumor_vol_pivot.index, avg_tumor_vol_pivot['Placebo'], yerr=sem_tumor_vol_pivot['Placebo'], color="y", marker="1", linewidth=0.3, ms=4)
#plt.errorbar(avg_tumor_vol_pivot.index, avg_tumor_vol_pivot['Propriva'], yerr=sem_tumor_vol_pivot['Propriva'], color="k", marker="2", linewidth=0.3, ms=4)

plt.title('Tumor Response to Treatment')
plt.xlabel('Time')
plt.ylabel('Tumor Volume (mm3)')
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc='best', fontsize='small', fancybox=True)
#plt.legend(['Capomulin', 'Ceftamin', 'Infubinol', 'Ketapril', 'Naftisol', 'Placebo', 'Propriva'], loc='best')
plt.grid()

# Save the Figure
plt.savefig('Tumor Response to Treatment.png')

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_count = main.groupby(['Drug', 'Timepoint']).count()['Tumor Volume (mm3)']

# Convert to DataFrame
survival_count = pd.DataFrame({'Mouse Count': survival_count}).reset_index()

# Preview DataFrame
survival_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_count_pivot = survival_count.pivot(index='Timepoint', columns='Drug')['Mouse Count']

# Preview the Data Frame
survival_count_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot((survival_count_pivot['Capomulin'] / 25) * 100, color="b", marker="o", linestyle='dashed', ms=4, linewidth=0.55)
plt.plot((survival_count_pivot['Infubinol'] / 25) * 100, color="r", marker="^", linestyle='dashed', ms=4, linewidth=0.55)
plt.plot((survival_count_pivot['Ketapril'] / 25) * 100, color="c", marker="<", linestyle='dashed', ms=4, linewidth=0.55)
plt.plot((survival_count_pivot['Placebo'] / 25) * 100, color="y", marker="1", linestyle='dashed', ms=7, linewidth=0.55)
plt.title('Mouse Survival Rates')
plt.ylabel('Survival Rate (%)')
plt.xlabel('Time')
plt.grid()
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc='best', fontsize='small', fancybox=True)

# Save the Figure
plt.savefig('Mouse Survival Rates.png')

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_change =  ((avg_tumor_vol_pivot.iloc[-1] - avg_tumor_vol_pivot.iloc[0]) / avg_tumor_vol_pivot.iloc[0]) * 100
sem_tumor_change =  ((sem_tumor_vol_pivot.iloc[-1] - sem_tumor_vol_pivot.iloc[0]) / sem_tumor_vol_pivot.iloc[0]) * 100

# Display the data to confirm
tumor_change
```

```python
# Store all Relevant Percent Changes into a Tuple
relevant_changes = (tumor_change['Capomulin'], 
                    tumor_change['Infubinol'], 
                    tumor_change['Ketapril'], 
                    tumor_change['Placebo'])

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
ind = np.arange(len(relevant_changes))  
width = 1
drug_pass = ax.bar(ind[0], relevant_changes[0], width, color='green')
drug_fail = ax.bar(ind[1:], relevant_changes[1:], width, color='red')

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_autoscaley_on(False)
ax.set_ylim([-30,70])
ax.grid()

# Use functions to label the percentages of changes
def labelPass(p):
    for p in p:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width()/2., -8, '-%d%% ' % int(height),
                ha='center', va='bottom', color='white')
def labelFail(f):
    for f in f:
        height = f.get_height()
        ax.text(f.get_x() + f.get_width()/2., 3, '%d%%' % int(height),
                ha='center', va='bottom', color='white')

# Call functions to implement the function calls
labelPass(drug_pass)
labelFail(drug_fail)

# Save the Figure
fig.savefig("Tumor Change.png")

# Show the Figure
fig.show()
```
